```python
# We'd like to know how many aerators were rented out on a given day.
# Since we want to join the Aerator csv with the Dates csv in SQL, we'll use this notebook to 
# create appropriate fields for rental start and end.
import pandas as pd
import numpy as np
import seaborn as sns

a_df = pd.read_csv('Aerator.csv')
a_df.head()
```

```python
# Looks like our 'Rental start' and 'Rental end' columns are still objects and not in datetime format.
a_df.info()
```

```python
# Let's take just the part we want, in string form, and move it to new columns.
# We'll drop the originals and the timestamps, too.
a_df[['rental_start', 'rt1']] = a_df['Rental start'].str.split(' ', n=1, expand=True)
a_df[['rental_end', 'rt2']] = a_df['Rental end'].str.split(' ', n=1, expand=True)
a_df = a_df.drop(['rt1', 'rt2', 'Rental start', 'Rental end'], axis=1)

a_df.head()
```

```python
# Now we'll convert them to datetime and make sure they're the correct format.
a_df['rental_start'] = pd.to_datetime(a_df['rental_start'])
a_df['rental_end'] = pd.to_datetime(a_df['rental_end'])
a_df.info()
```

```python
# Let's take a quick peek at the distribution of aerator rentals over time.
sns.histplot(a_df, x='rental_start')
```

```python
# Ok, I've created a csv of every date included in the all_rentals csv.
# My intention is to use it to track how many aerators were rented out per day.
dates_df = pd.read_csv('Dates - Sheet1.csv')
dates_df.head()
```

```python
# Let's make sure 'Date' is in datetime format.
dates_df['Date'] = pd.to_datetime(dates_df['Date'])
dates_df.head()
```

```python
# And now let's format it in the same way as the start and end dates in the aerator df.
dates_df['Date'] = pd.to_datetime(dates_df['Date'], format='%YYYY-%MM-%DD')
dates_df.head()
```

```python
# Let's add a field for aerator inventory and start it off at 0.
dates_df['aer_inv'] = 0
dates_df.head()
```

```python
# Now we'll merge the two dataframes in a left join, so all of the dates in dates_df are represented.
dates_df = dates_df.merge(a_df, how='left', left_on='Date', right_on='rental_start')
dates_df.head()
```

```python
pd.date_range(start=a_df['rental_start'][0], end=a_df['rental_end'][0])
# DatetimeIndex(['2018-01-01', '2018-01-02', '2018-01-03', '2018-01-04',
#                '2018-01-05', '2018-01-06', '2018-01-07', '2018-01-08'],
#               dtype='datetime64[ns]', freq='D')
```
